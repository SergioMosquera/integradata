**Integracion de datos**

A continuacion vamos a visualizar algunos ejemplos o tecnicas que permiten unir conjuntos de datos.

```python
import pandas as pd
wine_red='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/winequality-red.csv'
wine_white='https://raw.githubusercontent.com/terranigmark/curso-analisis-exploratorio-datos-platzi/main/winequality-white.csv'
```

```python
wred=pd.read_csv(wine_red, delimiter=';')
wwhite=pd.read_csv(wine_white, delimiter=';')
wred.head()
```

```python
wwhite.head()
```

Antes de unir cualquier **dataset** debes entender de donde vienen estos datos, por que lo vas a unir y como los vas a unir.

Daremos un distintivo a nuestros datos para conocer el conjunto original al cual pertenecen.

---



```python
wred['color']= 'red'
wwhite['color'] = 'white'
```

```python
wines=wred.append(wwhite)
wines
```

**CONCAT**

```python
pd.concat([wwhite,wred])
```

**UNION DE DATAFRAMES**

la union de data-frames por datakey iguales



```python
left=pd.DataFrame({'Key':['K1','K2','K3'],'A':['a1','a2','a3']})
right=pd.DataFrame({'Key':['K1','K4','K5'],'B':['b4','b5','b6']})

pd.merge(left,right,on='Key')
```

LEFT JOIN

```python
pd.merge(left,right,on='Key', how='left')
```
